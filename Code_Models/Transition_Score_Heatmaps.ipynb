```python
import numpy as np
import pickle
from matplotlib import pyplot as plt
```

### Transition score heatmaps

```python
# heat map tau & beta aggregation score

with open('final_files/agg_score_beta_das_LM_ML', 'rb') as fp:
                        agg_score_beta_das_LM_ML = pickle.load(fp)
with open('final_files/agg_score_tau_das_LM_ML', 'rb') as fp:
                        agg_score_tau_das_LM_ML = pickle.load(fp)

with open('final_files/agg_score_beta_da_LM_ML', 'rb') as fp:
                        agg_score_beta_da_LM_ML = pickle.load(fp)
with open('final_files/agg_score_tau_da_LM_ML', 'rb') as fp:
                        agg_score_tau_da_LM_ML = pickle.load(fp)


models = ['+Droplet drivers \n & +Amyloids\n vs. -PT', '+Droplet drivers \n vs. +Amyloids']   
#proteins = ['Amyloid-beta \n precursor', 'tau']
proteins = ['amyloid beta \n precursor protein', 'microtubule associated \n protein tau']
agg_score = np.round(np.array([[agg_score_beta_das_LM_ML['das'], agg_score_beta_da_LM_ML['da']],
                      [agg_score_tau_das_LM_ML['das'], agg_score_tau_da_LM_ML['da']]]), decimals=3)



fig, ax = plt.subplots(figsize=(8.8,8.8))
im = ax.imshow(agg_score, cmap = 'bwr', alpha=0.4)

# Create colorbar
cbar_kw = {}
cbarlabel="Transition score"
cbar = ax.figure.colorbar(im, ax=ax, **cbar_kw)
cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom", fontsize=20)
cbar.ax.tick_params(labelsize=20)



# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(models)), labels=models, fontsize=20)
ax.set_yticks(np.arange(len(proteins)), labels=proteins, fontsize=20)


# Rotate the tick labels and set their alignment.
#plt.setp(ax.get_xticklabels(), rotation=0, ha="right",
         #rotation_mode="anchor")
ax.grid(False)

text1 = np.array([['+Droplet drivers &\n +Amyloids', '+Amyloids'], ['+Droplet drivers &\n +Amyloids', '+Droplet drivers']])

# Loop over data dimensions and create text annotations.
for i in range(len(proteins)):
    for j in range(len(models)):
        #text = ax.text(j, i, f'{agg_score[i, j]}\n\n{text1[j]}',
         text = ax.text(j, i, f'\n\n{text1[i,j]}',
                       ha="center", va="center", color="black", fontsize=20)
```
